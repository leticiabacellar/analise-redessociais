```python
# Importando bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um conjunto de dados fictício de atividade em redes sociais
dados_redes_sociais = {
    'Usuario': ['Usuario1', 'Usuario2', 'Usuario3', 'Usuario4', 'Usuario5'],
    'Seguidores': [1500, 2000, 3000, 1200, 2500],
    'Curtidas': [200, 350, 150, 400, 300],
    'Comentarios': [15, 25, 10, 30, 20],
    'Compartilhamentos': [10, 8, 5, 15, 12]
}

# Criando um DataFrame a partir do conjunto de dados
df_redes_sociais = pd.DataFrame(dados_redes_sociais)

# Visualizando os primeiros registros do DataFrame
print(df_redes_sociais.head())

# Análise de Engajamento por Usuário
df_redes_sociais['Engajamento'] = df_redes_sociais['Curtidas'] + df_redes_sociais['Comentarios'] + df_redes_sociais['Compartilhamentos']

# Visualizando gráfico de engajamento por usuário
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x='Usuario', y='Engajamento', data=df_redes_sociais, palette='viridis')
ax.set_title('Engajamento por Usuário')
ax.set_ylabel('Engajamento')
plt.show()

# Análise de Popularidade de Postagens
df_redes_sociais['Popularidade'] = df_redes_sociais['Curtidas'] + 2 * df_redes_sociais['Comentarios'] + 3 * df_redes_sociais['Compartilhamentos']

# Visualizando gráfico de popularidade de postagens
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x='Usuario', y='Popularidade', data=df_redes_sociais, palette='viridis')
ax.set_title('Popularidade de Postagens por Usuário')
ax.set_ylabel('Popularidade')
plt.show()
```
